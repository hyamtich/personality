```python
#function to extract only user written text from the user timeline
def get_user_text(screenname):
    public_tweets = api.user_timeline(screen_name=screenname)
    user_text = []

    # foreach through all tweets pulled
    for tweet in public_tweets:
        if tweet.text.startswith("RT"):
            continue
        replaced = re.sub('http\S+','', tweet.text) 
        replaced = re.sub('@\S+','', replaced)
        #remove quotes
        replaced = re.sub('\"\S+\"','',replaced)
        user_text.append(replaced)
    
    return(user_text)
    
```
